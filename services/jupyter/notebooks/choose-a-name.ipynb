```python
import os
import geopandas as gpd
from infdb import InfDB
```

## Initialization

```python
# Initialize InfDB handler
infdb = InfDB(tool_name="choose-a-name", config_path="config.yml")

# Initialize logger
log = infdb.get_logger()

# Database connection
infdb_client = infdb.connect()
infdb_engine = infdb.get_db_engine()

# Start message
log.info(f"Starting {infdb.get_toolname()} tool")
```

## Read Data from InfDB

```python
# Read building data from InfDB database
# Hint: LOD2 data needs to be imported first -> configs/config-infdb-import.yml -> bash infdb.sh import
sql = "SELECT * FROM opendata.building_view LIMIT 100"
with infdb_engine.connect() as conn:
    gdf_buildings = gpd.read_postgis(sql, conn)
gdf_buildings.head()
```

```python
gdf_buildings.explore()
```

## Execute SQL Query

```python
# Schema configuration
format_params = {
    "output_schema": infdb.get_config_value([infdb.get_toolname(), "data", "output_schema"]),
}

# Drop output schema if exists for development purposes
log.info("Running SQL query ...")
infdb_client.execute_query("DROP SCHEMA IF EXISTS {output_schema} CASCADE".format(**format_params))
```

## Execute SQL Files

```python
# Schema configuration
format_params = {
    "output_schema": infdb.get_config_value([infdb.get_toolname(), "data", "output_schema"]),
}

# Execute sql scripts
log.info("Running SQL scripts ...")
SQL_DIR = os.path.join("sql")  # add subfolders here if needed
infdb_client.execute_sql_files(SQL_DIR, format_params=format_params)
```
